```python
df.drop(['EMAIL','CERTIFICATION_LINK', 'EMP_NAME', 'DESCRIPTION'],axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
df['SKILL_STATUS'] = df['SKILL_STATUS'].astype(str)
df['PROJECT_STATUS'] = df['PROJECT_STATUS'].astype(str)

# Define the mapping
mapping = {'Approved': 1, 'Not Approved': 0}

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply the mapping using the map() function
df['SKILL_STATUS'] = label_encoder.fit_transform(df['SKILL_STATUS'])
df['PROJECT_STATUS'] = label_encoder.fit_transform(df['PROJECT_STATUS'])
df
```

```python
df['SKILL_STATUS'] = df['SKILL_STATUS'].apply(lambda x: 1 if x == 0 else 0)
df['PROJECT_STATUS'] = df['PROJECT_STATUS'].apply(lambda x: 1 if x == 0 else 0)
# Display the DataFrame
df
```

```python
df2  = df
from sklearn.preprocessing import LabelEncoder
#df['PROFICIENCY'] = df['PROFICIENCY'].astype(str)
df2['SKILL_TECH'] = df2['SKILL_TECH'].astype(str)
df2['PROJECT_NAME'] = df2['PROJECT_NAME'].astype(str)
df2['PROJECT_TECH'] = df2['PROJECT_TECH'].astype(str)
df2['DESIGNATION'] = df2['DESIGNATION'].astype(str)

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply the mapping using the map() function
#df['PROFICIENCY'] = label_encoder.fit_transform(df['PROFICIENCY'])
df2['SKILL_TECH'] = label_encoder.fit_transform(df2['SKILL_TECH'])
df2['PROJECT_NAME'] = label_encoder.fit_transform(df2['PROJECT_NAME'])
df2['PROJECT_TECH'] = label_encoder.fit_transform(df2['PROJECT_TECH'])
df2['DESIGNATION'] = label_encoder.fit_transform(df2['DESIGNATION'])
df2
```

```python
# Create a new column 'Allocation' and initialize with zeros
df['Allocation'] = 1

# Apply conditions
condition = (df['PROFICIENCY'] == 1) & (df['SKILL_STATUS'] == "Approved") | (df['PROFICIENCY'] == 2)
grouped = df.groupby('EMP_ID')['PROJECT_NAME'].count()
user_ids_with_projects = grouped[grouped >= 1].index

# Set 'Allocation' to 1 where conditions are met
df.loc[condition & df['EMP_ID'].isin(user_ids_with_projects), 'Allocation'] = 0
df.head(25)
```

```python
desig_counts = df['Allocation'].value_counts()
plt.figure(figsize=(8, 4))  # Adjust the figure size if needed
desig_counts.plot(kind='bar', color='purple')

# Set the title and labels
plt.title('Distribution of Designation in the organization')
plt.xlabel('Designation')
plt.ylabel('Frequency')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
